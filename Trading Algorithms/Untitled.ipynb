```python
import feedparser

# URL of the RSS feed
rss_feed_url = 'https://rss.nytimes.com/services/xml/rss/nyt/Business.xml'

# Parse the RSS feed
feed = feedparser.parse(rss_feed_url)

# Print the title and link of each news item
print("Latest stock market news:")
for entry in feed.entries:
    print(f"Title: {entry.title}")
    print(f"Link: {entry.link}")
    print()
```

```python
from selenium import webdriver
from bs4 import BeautifulSoup

# Function to fetch news articles from a website
def fetch_news(url):
    driver = webdriver.Chrome()  # Assuming you have Chrome driver installed
    driver.get(url)
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    driver.quit()
    # Find elements containing news articles
    articles = soup.find_all('article')
    return articles

# Function to filter relevant news articles
def filter_articles(articles, stock_name):
    relevant_articles = []
    for article in articles:
        # Check if the article mentions the stock name
        if stock_name.lower() in article.text.lower():
            relevant_articles.append(article.text)
    return relevant_articles

# URL of the financial news website
url = 'https://www.google.com/finance/?hl=en'

# Stock name
stock_name = 'Apple'  # Change this to the stock you're interested in

# Fetch news articles
articles = fetch_news(url)

# Filter relevant articles
relevant_articles = filter_articles(articles, stock_name)

# Display relevant articles
print("Relevant articles about", stock_name)
for article in relevant_articles:
    print(article)
    print('-' * 50)
```
